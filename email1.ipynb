#### Cleaning

```python
# Deleting  the unnessary column in the dataset 
df.drop(['Email No.'],axis=1,inplace=True)
X = df.drop(['Prediction'],axis = 1)
y = df['Prediction']
```

```python
df.isnull().sum()
```

```python
#provides a collection of functions that make matplotlib work 
import matplotlib.pyplot as plt
%matplotlib inline 

#provides functions to control how warnings are issued and displayed.
import warnings
warnings.filterwarnings('ignore')#  tell to ignore all warnings that occur in the code.

#For training and testing that we importing this package.
from sklearn.model_selection import train_test_split

# This model is use for classification task in ml
from sklearn.svm import SVC

#It is use for evaluating the ML accuracy : accuracy score, confusion ,matrix etc
from sklearn import metrics
```

```python
# scalimg and standarizing your data
from sklearn.preprocessing import scale
X = scale(X)
```

```python
# Split data into train and testing 
#random number  = 42
X_train, X_test, y_train ,y_test = train_test_split(X, y,test_size=0.3, random_state=42)
```

```python
#classifies data points based on majority class among the nearest neighbours
from sklearn.neighbors import KNeighborsClassifier

#considering the lael of nearest 7 neighbours
knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
print("Prediction", y_pred)
```

```python
y_pred
```

```python
#predicting the accuracy
print("KNN accuracy = ",metrics.accuracy_score(y_test,y_pred))
print("Confusion matrix",metrics.confusion_matrix(y_test,y_pred))
```

```python

# svm separates data points in differrent classes in the feature space
model = SVC(C = 1)

# fitting the model
model.fit(X_train, y_train)

# predict
y_pred = model.predict(X_test)

metrics.confusion_matrix(y_true=y_test, y_pred=y_pred)
```

```python
print(" SVM Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
